```python
#check the df.iloc[i] =! reccomed[[i]] if you use concat or append
for i in range(len(df)):
    if df.iloc[i]["status"] == "close" and df.iloc[i+1]["status"] == "open":
        #print(str(i))
        d = pd.concat([df.iloc[i],df.iloc[i+1]])
        d.to_csv(str(i)+".csv")
```

```python
import glob
path = "preset/" # use your path
all_files = glob.glob(path + "*.csv")
frame = pd.DataFrame()
list_ = []
for file_ in all_files:
    df = pd.read_csv(file_,index_col=None, header=0)
    list_.append(df)
frame = pd.concat(list_)
```
